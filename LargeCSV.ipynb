```python
import dask.dataframe as dd
from dask.diagnostics import ProgressBar
```

# MÃ©todo de busca

```python
def CSV2DF(arquivos:str,colunas:list):
    return dd.read_csv(arquivos,low_memory=True, header=None,encoding='latin1',
             sep=";", names=colunas, dtype=str)
```

```python
def CSV_BUSCA(dataframe:dd,coluna:str,lista:list):
    with ProgressBar():
            return dataframe[dataframe[coluna].map(lambda x: x in lista)].compute()
```

# SOCIOS

```python
colunas=["CNPJ_BASICO","ID_TIPO_SOCIO","NOME_OU_RAZAO_SOCIAL","CNPJ_CPF","QUALIF_SOCIO",
         "DT_ENTRADA","CD_PAIS","REPR_LEGAL","NM_REPR","CD_QUALIF_REPR","FAIXA_ETARIA"]
pasta='SOCIOunzip\*'
df=CSV2DF(pasta,colunas)
df.head()
```
